## Арифметические операции

#### Задание 1 (Чаевые в ресторане)

**На вход:** сумма счёта (целое число, например, 2350) и процент чаевых, который вы хотите оставить (целое число, например, 15).

**На выходе:** итоговая сумма к оплате (счёт + чаевые) и сумма чаевых в рублях.

Формат вывода: `Счёт: 2350 руб. Чаевые (15%): 352.5 руб. Итого: 2702.5 руб.`


#### Задание 2 (Площадь кольца)

**На вход:** два целых числа: внешний радиус (R) и внутренний радиус (r) кольца.

**На выходе:** строка с данными и результатом вычисления площади кольца (S = π*(R² - r²)).

Формат вывода: `Площадь кольца с радиусами 5 и 3 = 50.24`

````python
def main():
    money = int(input())
    procent = int(input()) 

    tips = procent / 100 * money
    total = tips + money

    print(f"Счёт: {money} руб. Чаевые ({procent}%): {tips} руб. Итого: {total} руб.")

if __name__ == "__main__":
    main()
````

````python
from math import pi

def main():
    R = int(input())
    r = int(input())

    S = pi * (R ** 2 - r ** 2)
    print(f"Площадь кольца с радиусами {R} и {r} = {S:.2f}")

if __name__ == "__main__":
    main()
````

#### Задание 1 (Пропуск на мероприятие 18+)

**На вход:** получить из ввода три значения через input() (каждое на отдельной строке):

    Возраст посетителя (целое число).

    Наличие паспорта (строка "да" или "нет").

    Наличие пригласительного (строка "да" или "нет").

**На выходе:** Строка 'Проходите' или 'Вход воспрещён'.

**Условия:** Пройти можно, если:

    Возраст 18 или больше И есть паспорт.

    ИЛИ есть пригласительный (пригласительный отменяет требование возраста и паспорта).


````python
def main():
    age = int(input())
    have_passport = input()
    have_invitation = input()

    if (age >= 18 and have_passport == "да") or have_invitation == "да":
        print("Проходите")
    else:
        print("Вход воспрещён")

if __name__ == "__main__":
    main()
````

Пользователь вводит имя и фамилию через пробел. Программа генерирует никнейм.

#### Задание 2 (Поиск хэштега)

**На вход:** строка текста, в которой спрятан хэштег (слово, начинающееся с символа #, например, #путешествия).

Гарантируется, что в строке есть ровно один хэштег и он отделён пробелами от других слов.

Формат ввода: `Смотрите мои фото из #отпуск на море`

**На выход:** вывести только этот хэштег, но без символа # и в верхнем регистре.

Формат вывода: `ОТПУСК`

````python
def main():
    name = input()
    second_name = input()

    user = name[:3].upper() + "_" + second_name[:2].lower()
    len_name = len(name) + len(second_name) + 1

    print(user + str(len_name))

if __name__ == "__main__":
    main()
````

````python
def main():
    string = input()

    start_index = string.find("#")
    stop_index = string.find(" ", start_index)

    tag = string[start_index + 1:stop_index].upper()
    print(tag)

if __name__ == "__main__":
    main()
````

#### Задание 2 ("Угадай число")

Программа "загадывает" число от 1 до 100 (можно использовать переменную, например secret = 42).
Пользователю дается 5 попыток, чтобы угадать.

**На вход:** В цикле вводятся числа.

**На выходе:** После каждой попытки программа подсказывает "Больше" или "Меньше". Если число угадано — выводится "Поздравляю!" и цикл прерывается. Если попытки закончились — выводится "Попытки кончились. Было загадано: 42".

####  Задание 3 (Сумма цифр числа)

**На вход:** целое неотрицательное число.

**На выходе:** сумма его цифр.

**Условие:** Строки использовать нельзя.

**Пример:** 

    Ввод: 1234
    
    Вывод: 10

````python
def main():
    secret = 67
    count = 0
    guessed_right = False

    while count <= 5:
        x = int(input())

        if x == secret:
            guessed_right = True
            print("Поздравляю!")
            break

        count += 1

    if not guessed_right:
        print(f"Попытки кончились. Было загадано: {secret}")

if __name__ == "__main__":
    main()
````

````python
def count(x):
    "Сумма цифр числа"

    sum = 0

    while x > 0:
        sum += x % 10
        x //= 10

    return sum

def main():
    x = int(input())
    sum = count(x)
    print(sum)

if __name__ == "__main__":
    main()
````
